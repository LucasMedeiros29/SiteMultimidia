```
import cv2 as cv
import os
image_path = "/content/drive/My Drive/Colab Notebooks/"
img = cv.imread(os.path.join(image_path,'messi5.jpg'),cv.IMREAD_COLOR)
# BGR
plt.imshow(img)
plt.show()
```

```
# Para termos RGB invertem-se a ordem das cores:
imgRGB = img[:,:,::-1]
plt.imshow(imgRGB)
plt.show()
```

# O OpenCV tem uma função para isso, cvtColor():


```
# BGR to RGB
imgRGB = cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(imgRGB)
plt.show()
```

```
img_gray = cv.imread(os.path.join(image_path,'grupo_avatar.jpg'), cv.IMREAD_GRAYSCALE) #lê a imagem e converte para a escala de cinza
plt.imshow(img_gray,'gray') #exibe a imagem
plt.show()

```

# Há uma versão do google para o colab, do imshow:


```
from google.colab.patches import cv2_imshow # versão do Colab para o comando acima
cv2_imshow(img_gray)
```

Podemos usar resize() do OpenCV para mudar a resolução da imagem, façamos com a nossa P&B:

```
print(img_gray.shape) #exibe a dimensão da imagem
img2 = cv.resize(img_gray, None, fx=1.25, fy=1.25, interpolation = cv.INTER_CUBIC) #realiza o dimensionamento da imagem
print(img2.shape)
```

```
cv2_imshow(img2)
```

# 3. REPRODUÇÃO DE VÍDEO

<h2> Podemos usar o HTML para exibir vídeos no Colab!

```

from IPython.display import HTML
from base64 import b64encode
video_path = "/content/drive/My Drive/Colab Notebooks/Pikachu.mp4"
mp4 = open(video_path,'rb').read()
video_decodificado = "data:video/mp4;base64," + b64encode(mp4).decode()
HTML(f'<video width=720 controls><source src= {video_decodificado} type="video/mp4"></video>')

```
